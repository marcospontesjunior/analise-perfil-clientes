### Bibliotecas

```python
import pandas as pd
import plotly.express as px
```

### Importando e Lendo Base de Dados

```python
dados = pd.read_csv('clientes.csv', sep=';', encoding='latin')
dados.head(10)
```

### Tratando e Limpando a Base de Dados

```python
display(dados.info())
```

```python
# tratamento

dados['Salário Anual (R$)'] = pd.to_numeric(dados['Salário Anual (R$)'], errors='coerce')
```

```python
# limpando

dados = dados.dropna()

dados = dados.drop('Unnamed: 8', axis=1)
```

```python
dados.head(10)
```

```python
display(dados.info())
```

### Analisando Dados

```python
display(dados.describe())
```

```python
for colunas in dados.columns:
    grafico = px.histogram(dados, x=colunas, y='Nota (1-100)', histfunc='avg', text_auto=True, nbins=5)
    grafico.show()
```
